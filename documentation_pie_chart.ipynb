# Pie Chart Documentation

List of parameters for the musical_pie_plot function:
- location: direct path to the datafile
- dataType: type of data that the program has to read
- log: boolean that determines if the color scale is logarithmic or linear
- convertTable: if the dataType to be read is pitch class, provides a dictionary to the tpc format
- pitchClassDisplay: boolean that determines if the pie use the pitch class format or the tpc format to represent the notes
- colorbar: boolean that determines of it has to plot the colorbar or note
- duration: boolean that determines if it has to account of the total duration of the note or the number of time the note appeared
- fifth: boolean that determines if the note are placed side by side as fifth or chromatic
- figSize: give the size of the plot in inches
- topNote: tell the note that has to be on top
- rotation: if you want you can rotate the graph in the trigonometric wise in degres
- clockwise: booleans taht determines of the note rotate clockwise or counter clockwise
- colorGeneral: tell the generalt color or color panel for the graph
- colorZeroValue: tell the color of the notes that are not represented
<br>
<br>
